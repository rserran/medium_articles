# A simple way to download financial data from Investing.com in Python

## Setup

```python
import investpy as inv
```

## Stocks data

```python
inv.get_stocks_overview(country="United States", 
                        as_json=False, 
                        n_results=1000)
```

```python
TICKER = "AAPL"
```

```python
inv.get_stock_company_profile(stock=TICKER, country="United States")
```

```python
inv.get_stock_financial_summary(stock=TICKER,
                                country="United States",
                                summary_type="income_statement", 
                                period="quarterly")
```

```python
inv.get_stock_information(stock=TICKER,
                          country="United States",
                          as_json=False).transpose()
```

```python
df = inv.get_stock_historical_data(stock=TICKER,
                                   country="United States",
                                   from_date="01/01/2020",
                                   to_date="01/01/2021")
df
```

```python
aapl = inv.search_quotes(text="apple", 
                         products=["stocks"],
                         countries=["united states"], 
                         n_results=1)
print(aapl)
```

```python
recent_data = aapl.retrieve_recent_data()
recent_data
```

```python
historical_data = aapl.retrieve_historical_data(from_date="01/01/2020", 
                                                to_date="01/01/2021")
historical_data
```

```python
aapl.retrieve_technical_indicators(interval="daily")
```

```python
aapl.retrieve_technical_indicators(interval="weekly")
```

## Crypto data

```python
inv.search_cryptos(by="name", value="doge")
```

```python
df = inv.get_crypto_historical_data(crypto="Dogecoin",
                                    from_date="01/01/2020",
                                    to_date="11/01/2020")
df
```

```python
search_result = inv.search_quotes(text="doge", 
                                  products=["cryptos"],
                                  n_results=100)
for res in search_result:
    print(res)
```

## Bond data

```python
inv.get_bond_historical_data(bond="Netherlands 10Y", 
                             from_date="01/01/2020", 
                             to_date="01/12/2020")
```
